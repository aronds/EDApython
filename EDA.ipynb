## Creamos un dataset ficticio

```python
fake = Faker()
geolocator = Nominatim(user_agent="my_app")

# Crear un diccionario con los nombres de las columnas
columnas = {
    'id_propiedad': [],
    'tipo_propiedad': [],
    'ubicacion': [],
    'latitud': [],
    'longitud': [],
    'precio': [],
    'superficie': [],
    'numero_habitaciones': [],
    'caracteristicas_adicionales': [],
    'ano_construccion': [],
    'estado_propiedad': [],
    'agencia_inmobiliaria': [],
    'fecha_publicacion': [],
    'descripcion': []
}

# Generar 1000 registros
for _ in range(1000):
    id_propiedad = fake.random_number(digits=6)
    tipo_propiedad = fake.random_element(elements=('Casa', 'Apartamento', 'Oficina', 'Local comercial'))
    ubicacion = fake.city()
    location = geolocator.geocode(ubicacion)
    
    if location is not None:
        latitud = location.latitude
        longitud = location.longitude
    else:
        latitud = None
        longitud = None
    
    precio = fake.random_int(min=100000, max=1000000)
    superficie = fake.random_int(min=50, max=500)
    numero_habitaciones = fake.random_int(min=1, max=5)
    caracteristicas_adicionales = fake.random_elements(elements=('Piscina', 'Jardín', 'Terraza', 'Cochera'), unique=True)
    ano_construccion = fake.random_int(min=1970, max=2020)
    estado_propiedad = fake.random_element(elements=('Nuevo', 'Usado', 'En construcción'))
    agencia_inmobiliaria = fake.company()
    fecha_publicacion = fake.date_between(start_date='-1y', end_date='today')
    descripcion = fake.paragraph()

    columnas['id_propiedad'].append(id_propiedad)
    columnas['tipo_propiedad'].append(tipo_propiedad)
    columnas['ubicacion'].append(ubicacion)
    columnas['latitud'].append(latitud)
    columnas['longitud'].append(longitud)
    columnas['precio'].append(precio)
    columnas['superficie'].append(superficie)
    columnas['numero_habitaciones'].append(numero_habitaciones)
    columnas['caracteristicas_adicionales'].append(caracteristicas_adicionales)
    columnas['ano_construccion'].append(ano_construccion)
    columnas['estado_propiedad'].append(estado_propiedad)
    columnas['agencia_inmobiliaria'].append(agencia_inmobiliaria)
    columnas['fecha_publicacion'].append(fecha_publicacion)
    columnas['descripcion'].append(descripcion)

# Crear un DataFrame a partir del diccionario de columnas
df = pd.DataFrame(columnas)

# Mostrar el DataFrame
df
```
